```python
def inputfunction(x,n,find1):##设定输入函数/set input function
    x = input().split(" ")
    while(len(x)< n):
        print("数据数量不足，请重新输入数据/the number of the data is not enough, please reinput the data of x")
        x = []
        x = input().split(" ")##如果输入数量不足，要求重新输入，直至符合输入数量为止/If the input quantity is insufficient, 
                              ##re-input is required until the input quantity meets the requirement

    while(len(x) > n):
        print("数据数量过多，请重新输入数据/the number of the data is too many, please reinput the data of x")
        x = []
        x = input().split(" ")##如果输入数量过多，要求重新输入，直至符合输入数量为止/If the number of inputs is too high,
                              ## re-input is required until the input quantity meets the requirement

    for i in range(n):
        if(find1 in x[i]):
            x[i]= float(x[i])
        else:
            x[i]=int(float(x[i]))##因为输入数据为int或float格式，需要进行区分/Because the input data is in int or float format,
                                 ##need to differentiate
    return x

            


print("请输入数据个数N/please enter the number of data N")

n=int(input())


print("请输入x的数据/please input the data of x")
x=[]
x=inputfunction(x,n,".")##区别int和float的简单方法：寻找小数点/Look for a decimal point to identify an int or float

print("x的数据展示如下：/the data of x is shown down below:")

print(x)

print("请输入y的数据/please input the data of y")
y=[]
y=inputfunction(y,n,".")
        
print("y的数据展示如下：/the data of y is shown down below:")

print(y)




print("开始计算相似度，使用方法1:欧几里德公式/starting calculating the similarity,using measure1: Euclidean")

sumEu = float(0)

for i in range(n):
    sumEu= sumEu + (x[i]-y[i])*(x[i]-y[i])##欧几里德距离公式Euclidean distance formula
    
disEu=sumEu**0.5

print(disEu)



print("开始计算相似度，使用方法2:余弦公式/starting calculating the similarity,using measure2: cosine")


sumup = float(0)
sumdownx = float(0)
sumdowny = float(0)

for i in range(n):
    sumup = sumup + x[i]*y[i];
    sumdownx = sumdownx + x[i]*x[i]
    sumdowny = sumdowny + y[i]*y[i]
     
disCo=sumup/((sumdownx**0.5)*(sumdowny**0.5))##余弦距离公式Cosine distance formula

print(disCo)

print("开始计算相似度，使用方法3:Jaccard公式/starting calculating the similarity,using measure3: Jaccard")

checkbinary = 1
sumupJa = 0
sumdownJa = 0

for i in range(n):
    if((x[i]!=0 and x[i]!=1) or(y[i]!=0 and y[i]!=1)):
        print('undefined')
        checkbinary = 0
        break

if(checkbinary == 1):
    for i in range(n):
        if(x[i]==1 and y[i]==1):
            sumupJa = sumupJa + 1
            sumdownJa = sumdownJa +1
        elif(x[i]!= 0 or y[i]!= 0):
            sumdownJa = sumdownJa + 1
    disJa=sumupJa/sumdownJa        #Jaccard formula j=f11/(f10+f01+ff11)
    print(disJa)    

        
print("开始计算相似度，使用方法4:Correlation公式/starting calculating the similarity,using measure4: Correlation")

sumx = float(0)
sumy = float(0)

for i in range(n):
    sumx = sumx + x[i]
    sumy = sumy + y[i]

meanx = sumx/n
meany = sumy/n


a=float(0)
b=float(0)
c=float(0)

for i in range(n):
    a = a + round((x[i]-meanx)*(y[i]-meany),10)
    b = b + (x[i]-meanx)*(x[i]-meanx)
    c = c + (y[i]-meany)*(y[i]-meany)


  
covxy =round((a/(n-1)),5)
stdx =(b/(n-1))**0.5
stdy =(c/(n-1))**0.5

if(stdx==0 or stdy==0):##correlation formula cov(xy)/(stdx*stdy)
    print("undefined")
else:
    corrxy = covxy/(stdx*stdy)
    print(round(corrxy,2) )







        
        
    
       
    
        
    


 
```
